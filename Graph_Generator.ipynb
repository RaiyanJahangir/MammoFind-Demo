```python
from pyvis.network import Network
import pandas as pd
from pandas import DataFrame
import textwrap

def extract_numbers(input_string):
    numbers_only = ''.join(char for char in input_string if char.isdigit())
    result_integer = int(numbers_only)
    return result_integer

def wrap_text(text, width=50):
    wrapper = textwrap.TextWrapper(width=width, break_long_words=False)
    return '\n'.join(wrapper.wrap(text))

def wrap_text_withBR(text, width=40):
    wrapper = textwrap.TextWrapper(width=width, break_long_words=False)
    return '<br>'.join(wrapper.wrap(text))

def processing(df):
  for column in df.columns:
    df[column] = df[column].str.lstrip().str.rstrip()
    return df

filename = "Mammography_Dataset.xlsx"

MainSheet = pd.read_excel(filename, sheet_name='main', skiprows=0)
TaskDescriptionSheet = pd.read_excel(filename, sheet_name='TaskDescription', skiprows=0)
StatusDescriptionSheet = pd.read_excel(filename, sheet_name='StatusDescription', skiprows=0)
DatatypeDescriptionSheet = pd.read_excel(filename, sheet_name='DatatypeDescription', skiprows=0)
# FiletypeDescriptionSheet = pd.read_excel(filename, sheet_name='FiletypeDescription', skiprows=0)
# LocationDescriptionSheet = pd.read_excel(filename, sheet_name='LocationDescription', skiprows=0)


MainSheet = processing(MainSheet)
TaskDescriptionSheet = processing(TaskDescriptionSheet)
StatusDescriptionSheet = processing(StatusDescriptionSheet)
DatatypeDescriptionSheet = processing(DatatypeDescriptionSheet)
# FiletypeDescriptionSheet = processing(FiletypeDescriptionSheet)
# LocationDescriptionSheet = processing(LocationDescriptionSheet)

# Display the modified DataFrame
MainSheet['Information'] = MainSheet['Information'].apply(wrap_text)

Data_Dict = dict(zip(MainSheet['Name'], MainSheet['Information']))
TaskDescription_Dict = dict(zip(TaskDescriptionSheet['TaskName'], TaskDescriptionSheet['DescriptionOfTheTask']))
StatusDescription_Dict = dict(zip(StatusDescriptionSheet['StatusName'], StatusDescriptionSheet['DescriptionOfTheStatus']))
DatatypeDescription_Dict = dict(zip(DatatypeDescriptionSheet['DatatypeName'], DatatypeDescriptionSheet['DescriptionOfTheDatatype']))
# FiletypeDescription_Dict = dict(zip(FiletypeDescriptionSheet['FiletypeName'], FiletypeDescriptionSheet['DescriptionOfTheFiletype']))
# LocationDescription_Dict = dict(zip(LocationDescriptionSheet['LocationName'], LocationDescriptionSheet['DescriptionOfTheLocation']))

DataAndInfo = MainSheet[['Name', 'Information']]
MainSheet.head()
```

```python
DataAndTask = MainSheet[['Name', 'Tasks']]

b = DataFrame(DataAndTask.Tasks.str.split(',').tolist(), index=DataAndTask.Name).stack()
b = b.reset_index()[[0, 'Name']]
b.columns = ['Tasks', 'Name']
DataAndTask  = b
DataAndTask = processing(DataAndTask)
DataAndTask
```

```python
DataAndStatus = MainSheet[['Name', 'Status']]

b = DataFrame(DataAndStatus.Status.str.split(',').tolist(), index=DataAndStatus.Name).stack()
b = b.reset_index()[[0, 'Name']]
b.columns = ['Status', 'Name'] 
DataAndStatus  = b
DataAndStatus = processing(DataAndStatus)
DataAndStatus
```

```python
DataAndDatatype = MainSheet[['Name', 'Datatype']]

b = DataFrame(DataAndDatatype.Datatype.str.split(',').tolist(), index=DataAndDatatype.Name).stack()
b = b.reset_index()[[0, 'Name']]
b.columns = ['Datatype', 'Name'] 
DataAndDatatype  = b
DataAndDatatype = processing(DataAndDatatype)
DataAndDatatype
```

```python
DataAndFiletype = MainSheet[['Name', 'Filetype']]

b = DataFrame(DataAndFiletype.Filetype.str.split(',').tolist(), index=DataAndFiletype.Name).stack()
b = b.reset_index()[[0, 'Name']]
b.columns = ['Filetype', 'Name'] 
DataAndFiletype  = b
DataAndFiletype = processing(DataAndFiletype)
DataAndFiletype
```

```python
DataAndLocation = MainSheet[['Name', 'Location']]

b = DataFrame(DataAndLocation.Location.str.split(',').tolist(), index=DataAndLocation.Name).stack()
b = b.reset_index()[[0, 'Name']]
b.columns = ['Location', 'Name'] 
DataAndLocation  = b
DataAndLocation = processing(DataAndLocation)
DataAndLocation
```

```python
DataInfoTask = pd.merge(DataAndInfo, DataAndTask, on='Name', how='outer')
DataInfoTask['TaskDescription'] = None
DataAndReference = MainSheet[['Name', 'Availability', 'Articles','Filetype','Location']]

DataTaskDescription = DataInfoTask


for i in range(0,len(DataTaskDescription)):
 DataTaskDescription.TaskDescription[i] = TaskDescription_Dict[DataTaskDescription.Tasks[i].lstrip().rstrip()]


DataInfoTaskStatus = pd.merge(DataInfoTask, DataAndStatus, on='Name', how='outer')
DataInfoTaskStatus['StatusDescription'] = None

DataInfoTaskStatusDescription=DataInfoTaskStatus
for i in range(0,len(DataInfoTaskStatusDescription)):
 DataInfoTaskStatusDescription.StatusDescription[i] = StatusDescription_Dict[DataInfoTaskStatus.Status[i].lstrip().rstrip()]


DataInfoTaskStatusDatatype = pd.merge(DataInfoTaskStatus, DataAndDatatype, on='Name', how='outer')
DataInfoTaskStatusDatatype['DatatypeDescription'] = None

DataInfoTaskStatusDatatypeDescription=DataInfoTaskStatusDatatype
for i in range(0,len(DataInfoTaskStatusDatatypeDescription)):
 DataInfoTaskStatusDatatypeDescription.DatatypeDescription[i] = DatatypeDescription_Dict[DataInfoTaskStatusDatatype.Datatype[i].lstrip().rstrip()]



DataInfoTaskStatusDatatypeFiletype = pd.merge(DataInfoTaskStatusDatatype, DataAndFiletype, on='Name', how='outer')
DataInfoTaskStatusDatatypeFiletype['FiletypeDescription'] = None

DataInfoTaskStatusDatatypeFiletypeDescription=DataInfoTaskStatusDatatypeFiletype
for i in range(0,len(DataInfoTaskStatusDatatypeFiletypeDescription)):
 DataInfoTaskStatusDatatypeFiletypeDescription.FiletypeDescription[i] = FiletypeDescription_Dict[DataInfoTaskStatusDatatypeFiletype.Filetype[i].lstrip().rstrip()]



DataInfoTaskStatusDatatypeFiletypeLocation = pd.merge(DataInfoTaskStatusDatatypeFiletype, DataAndLocation, on='Name', how='outer')
DataInfoTaskStatusDatatypeFiletypeLocation['LocationDescription'] = None

DataInfoTaskStatusDatatypeFiletypeLocationDescription=DataInfoTaskStatusDatatypeFiletypeLocation
for i in range(0,len(DataInfoTaskStatusDatatypeFiletypeLocationDescription)):
 DataInfoTaskStatusDatatypeFiletypeLocationDescription.LocationDescription[i] = LocationDescription_Dict[DataInfoTaskStatusDatatypeFiletypeLocation.Location[i].lstrip().rstrip()]




Data_Info_Task_Status_Datatype_TaskDescription = DataInfoTaskStatusDatatype
Data_Info_Task_Status_Datatype_TaskDescription_Reference = pd.merge(Data_Info_Task_Status_Datatype_TaskDescription, DataAndReference, on='Name', how='outer')
Data_Info_Task_Status_Datatype_TaskDescription_Reference

#The following line is added to remove the exact duplicate rows based on ['Name', 'Tasks', 'Status','Datatype','Filetype','Location']
Data_Info_Task_Status_Datatype_TaskDescription_Reference = Data_Info_Task_Status_Datatype_TaskDescription_Reference.drop_duplicates(subset=['Name', 'Tasks', 'Status','Datatype','Filetype','Location'])
Data_Info_Task_Status_Datatype_TaskDescription_Reference.reset_index(inplace=True, drop=True)
Data_Info_Task_Status_Datatype_TaskDescription_Reference
```

```python
from pyvis.network import Network

# Create the network
net = Network(directed=False, height="100%", width="100%", physics=False)

# Define the fixed positions for certain nodes
fixed_positions = {
    'data_Name': {'x': 0, 'y': 0},
    'Task': {'x': 100, 'y': 100},
    'stat': {'x': 200, 'y': 0},
    'Datatype': {'x': 300, 'y': 100},
    # Add positions for other nodes if needed
}

# Add nodes to the network with fixed positions
for i in range(0, len(data_Name)):
    # Add the node for data_Name
    net.add_node(data_Name[i], label=data_Name[i], title=data_Info[i], color='#87CEEB', x=fixed_positions.get(data_Name[i], {}).get('x', 0), y=fixed_positions.get(data_Name[i], {}).get('y', 0))
    
    # Add the node for Task
    net.add_node(Task[i], label=Task[i], title=TaskDescription[i], color="#FFDF00", x=fixed_positions.get(Task[i], {}).get('x', 100), y=fixed_positions.get(Task[i], {}).get('y', 100))
    
    # Add the node for stat
    net.add_node(stat[i], label=stat[i], title=StatusDescription[i], color="#90EE90", x=fixed_positions.get(stat[i], {}).get('x', 200), y=fixed_positions.get(stat[i], {}).get('y', 0))
    
    # Add the node for Datatype
    net.add_node(Datatype[i], label=Datatype[i], title=DatatypeDescription[i], color="#FF7F7F", x=fixed_positions.get(Datatype[i], {}).get('x', 300), y=fixed_positions.get(Datatype[i], {}).get('y', 100))

    # Add the edges between nodes
    net.add_edge(data_Name[i], Task[i], weight=1, title=Edges[i], color="#898980")
    net.add_edge(data_Name[i], stat[i], weight=1, title=Edges[i], color="#898980")
    net.add_edge(data_Name[i], Datatype[i], weight=1, title=Filetypes[i], color="#898980")

# Add additional edges
net.add_edge('CBIS-DDSM', 'RBIS-DDSM', weight=1, title='RBIS is further curated and smaller version of CBIS-DDSM', color="#898980")
net.add_edge('CBIS-DDSM', 'DDSM', weight=1, title='CBIS-DDSM is an extended and modified version of DDSM', color="#898980")
net.add_edge('OPTIMAM', 'OMI-DB', weight=1, title='OPTIMAM is a newer version of OMI-DB', color="#898980")

# Generate the HTML file with the network graph
net.write_html('Sample_Graph.html')
```
